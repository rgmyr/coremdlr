```python
import numpy as np
import pywt

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
ascent = pywt.data.ascent()

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(ascent, cmap='gray')
```

```python
ecg = pywt.data.ecg()

fig, ax = plt.subplots(figsize=(15,5))
ax.plot(np.arange(ecg.size), ecg)
```

```python
morl = pywt.ContinuousWavelet('morl')
print(morl)
```

```python
ecg_cA, ecg_cD = pywt.dwt(ecg, 'db1', mode='symmetric', axis=-1)
```

```python
plt.plot(x=np.arange(ecg_cA.size), y = ecg_cA)
```
