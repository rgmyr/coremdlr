```python
import numpy as np
from skimage import io, morphology
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def generate_mask(img):
    '''Label color used is RGB (0,255,255)'''
    reds   = img[:, :, 0]
    greens = img[:, :, 1]
    blues  = img[:, :, 2]
    
    m1 = np.logical_and(greens==255, blues==255)

    return np.logical_and(m1, reds==0)
```

```python
f, ax = plt.subplots(figsize=(15,15))

img = io.imread('../data/core_photos/Shell_jpg/1643_51_4389-4393m_BlueLabel.jpg')
ax.imshow(img)
```

```python
f, ax = plt.subplots(figsize=(15,15))
ax.imshow((generate_mask(img)))
```
