```python
import numpy as np
import pandas as pd

import hyperopt
from hyperopt import hp
from hyperopt.pyll.base import scope

from corebreakout.facies.datasets import DepthwiseSequenceGenerator, FaciesDataset
```

```python
from tensorflow.keras.models import Model as KerasModel
from tensorflow.keras.layers import Conv1D, Input, Dense, Reshape, Dropout

def conv1d_net(num_classes,
               input_shape,
               filters=[32,48,64,128,256],
               output_resolution=32,
               activation='gated',
               dense_layers=[],
               dropout_rate=None):
    #CNN built on dilation blocks and auxillary poolings.

    input_signal = Input(shape=input_shape)
    x = Reshape((input_shape[0],1))(input_signal) if len(input_shape) <= 1 else input_signal
    
    x = Conv1D(filters[0], 3, strides=2, padding='same', activation='relu')(x)

    for f in filters[1:]:
        x = Conv1D(f, 3, strides=2, padding='same', activation='relu')(x)
        
    x = Dropout(dropout_rate)(x)

    preds = Dense(4, activation='softmax')(x)
    
    return KerasModel(inputs=input_signal, outputs=preds)

conv1dnet = conv1d_net(4, (32*256*4,), dropout_rate=0.5)
conv1dnet.summary()
```

```python
fdset = FaciesDataset(["205-21b-3", "204-19-6", "204-24a-6"],
                    test_wells=["204-20-6a"],
                    features=["pseudoGR"],
                    label_resolution=32)

fdset.load_or_generate_data()
```

```python
train_generator = DepthwiseSequenceGenerator(fdset.X_train['pseudoGR'], fdset.y_train, 256*4, step_size=64, 
                                             batch_size=8, concatenate_X=True)
test_generator = DepthwiseSequenceGenerator(fdset.X_test['pseudoGR'], fdset.y_test, 256*4, step_size=64, 
                                            batch_size=8, concatenate_X=True)

conv1dnet.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
hist =  conv1dnet.fit_generator(train_generator, epochs=100, validation_data=test_generator, shuffle=False)
```
