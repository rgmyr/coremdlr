```python
from tensorflow.keras import backend as K
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense, \
    Input, Lambda, Multiply, Add, Activation, Reshape

import os

os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="1"


def wavenetBlock(filters, residual_filters, kernel_size, dilation_rate):
    def f(input_):    
        residual = input_
        
        tanh_out = Conv1D(filters, kernel_size, dilation_rate=dilation_rate,
                          padding='same', activation='tanh')(input_)
        sigm_out = Conv1D(filters, kernel_size, dilation_rate=dilation_rate,
                          padding='same', activation='sigmoid')(input_)
        merged = Multiply()([tanh_out, sigm_out])
        
        skip_out = Conv1D(residual_filters, 1, activation='relu', padding='same')(merged)
        
        out = Add()([skip_out, residual])
        
        return out, skip_out
    
    return f


def wavenet(num_classes,
            input_size,
            block_filters=64,
            residual_filters=16,
            num_blocks=20,
            output_resolution=32):
    
    input_ = Input(shape=(input_size,))
    x = Reshape((input_size, 1))(input_)
    
    x = Conv1D(residual_filters, 7, padding='same', activation='relu')(x)
    
    A, B = wavenetBlock(block_filters, residual_filters, 3, 2)(x)
    skip_connections = [B]
    for i in range(num_blocks):
        A, B = wavenetBlock(block_filters, residual_filters, 3, 2**((i+2)%9))(A)
        skip_connections.append(B)
        
    net = Add()(skip_connections)
    net = Activation('relu')(net)
    
    net = Conv1D(residual_filters, 1, activation='relu')(net)
    net = Conv1D(residual_filters, output_resolution, strides=output_resolution, padding='valid')(net)
    #net = Reshape((-1, output_resolution))(net)
    net = Dense(num_classes, activation='softmax')(net)
    model = Model(inputs=input_, outputs=net)

    return model
```

```python
wnet = wavenet(4, 1024*4, num_blocks=20)
wnet.summary()
```

```python
from corebreakout.facies.datasets import DepthSequenceGenerator, FaciesDataset
```

```python
fdset = FaciesDataset(["205-21b-3", "204-20-6a", "204-24a-6"],
                    test_wells=["204-19-6"],
                    features=["pseudoGR"],
                    label_resolution=32)

fdset.load_or_generate_data()
```

```python
import hyperopt
from hyperopt import hp
from hyperopt.pyll.base import scope
from sklearn.metrics import f1_score, log_loss
from tensorflow.keras.optimizers import SGD, Adam

import pandas as pd
```

```python
WAVENET_SEARCH_SPACE = {
    'input_size' : scope.int(hp.quniform('input_size', 128, 512+256, 128)),
    'block_filters' : scope.int(hp.quniform('block_filters', 32, 128, 32)),
    'residual_filters' : scope.int(hp.quniform('residual_filters', 1, 64, 8)),
    'num_blocks' : scope.int(hp.quniform('num_blocks', 10, 30, 5)),
    'output_resolution' : 32
}
                       
results_df = pd.DataFrame(columns=list(WAVENET_SEARCH_SPACE.keys()) + ['best_epoch', 'val_loss', 'val_acc'])

def train_wavenet(model_config):
    global results_df
    
    print(model_config)
    input_size = model_config.pop('input_size')
    
    wnet = wavenet(4, input_size*32, **model_config)
    wnet.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    try:
        train_generator = DepthSequenceGenerator(fdset.X_train['pseudoGR'], fdset.y_train, input_size, 
                                             step_size=int(input_size//4), batch_size=6, concatenate_X=True)
        test_generator = DepthSequenceGenerator(fdset.X_test['pseudoGR'], fdset.y_test, input_size, 
                                            step_size=int(input_size//4), batch_size=6, concatenate_X=True)
        hist = wnet.fit_generator(train_generator, epochs=50, validation_data=test_generator, shuffle=False)
    # make batch_size smaller if too much memory
    except ResourceExhaustedError:
        train_generator = DepthSequenceGenerator(fdset.X_train['pseudoGR'], fdset.y_train, input_size, 
                                             step_size=int(input_size//4), batch_size=3, concatenate_X=True)
        test_generator = DepthSequenceGenerator(fdset.X_test['pseudoGR'], fdset.y_test, input_size, 
                                            step_size=int(input_size//4), batch_size=3, concatenate_X=True)
        hist = wnet.fit_generator(train_generator, epochs=50, validation_data=test_generator, shuffle=False)
    
    new_row = model_config
    new_row['best_epoch'] = np.argmin(hist.history['val_loss'])
    new_row['val_loss'] = min(hist.history['val_loss'])
    new_row['val_acc'] = max(hist.history['val_acc'])
    results_df = results_df.append(new_row, ignore_index=True)
    
    print('Best so far (loss, acc):', results_df['val_loss'].min(), results_df['val_acc'].max())
    
    K.clear_session()
    
    return new_row['val_loss']
```

```python
best_params = hyperopt.fmin(
    fn=train_wavenet,
    space=WAVENET_SEARCH_SPACE,
    algo=hyperopt.rand.suggest,
    max_evals=50
)
```

```python
print(results_df.sort_values('val_loss'))
results_df.to_csv('./pGR_wavenet_results.csv')
```
